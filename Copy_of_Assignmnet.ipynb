````python
# Aggregate Insights
total_sales = df["TransactionAmount"].sum()
total_quantity = df["Quantity"].sum()
average_transaction = df["TransactionAmount"].mean()
payment_method_distribution = df["PaymentMethod"].value_counts()
store_type_distribution = df["StoreType"].value_counts()
average_discount = df["DiscountPercent"].mean()
average_shipping_cost = df["ShippingCost"].mean()
average_delivery_time = df["DeliveryTimeDays"].mean()

# Drill-down Insights
city_sales = df.groupby("City")["TransactionAmount"].sum().sort_values(ascending=False)
region_sales = df.groupby("Region")["TransactionAmount"].sum().sort_values(ascending=False)
product_sales = df.groupby("ProductName")["TransactionAmount"].sum().sort_values(ascending=False)
returned_products = df[df["Returned"] == "Yes"].groupby("ProductName")["TransactionAmount"].count()
promotional_sales = df.groupby("IsPromotional")["TransactionAmount"].sum()

# Customer Insights
age_distribution = df["CustomerAge"].describe()
gender_distribution = df["CustomerGender"].value_counts()

# Print results
print("### Aggregate Insights ###")
print(f"Total Sales Revenue: â‚¹{total_sales}")
print(f"Total Quantity Sold: {total_quantity}")
print(f"Average Transaction Amount: â‚¹{average_transaction}")
print("\nPayment Method Distribution:\n", payment_method_distribution)
print("\nStore Type Distribution:\n", store_type_distribution)
print(f"\nAverage Discount Percentage: {average_discount}%")
print(f"Average Shipping Cost: â‚¹{average_shipping_cost}")
print(f"Average Delivery Time: {average_delivery_time} days")

print("\n### Drill-Down Insights ###")
print("\nSales by City:\n", city_sales)
print("\nSales by Region:\n", region_sales)
print("\nSales by Product:\n", product_sales)
print("\nReturned Products Count:\n", returned_products)
print("\nSales by Promotion:\n", promotional_sales)

print("\n### Customer Insights ###")
print("\nCustomer Age Distribution:\n", age_distribution)
print("\nCustomer Gender Distribution:\n", gender_distribution)
````

# SQL Scripts for Data Exploration and Key Insights

## 1. Total Sales Revenue
```sql
SELECT SUM(TransactionAmount) AS total_sales
FROM sales_data;

2. Total Quantity Sold
SELECT SUM(Quantity) AS total_quantity
FROM sales_data;

3. Average Transaction Amount
SELECT AVG(TransactionAmount) AS avg_transaction
FROM sales_data;

4. Payment Method Distribution
SELECT PaymentMethod, COUNT(*) AS payment_method_count
FROM sales_data
GROUP BY PaymentMethod
ORDER BY payment_method_count DESC;

5. Store Type Distribution
SELECT StoreType, COUNT(*) AS store_type_count
FROM sales_data
GROUP BY StoreType
ORDER BY store_type_count DESC;

6. Average Discount Percentage
SELECT AVG(DiscountPercent) AS avg_discount
FROM sales_data;

7.Average Shipping Cost
SELECT AVG(ShippingCost) AS avg_shipping_cost
FROM sales_data;

8.Average Delivery Time
SELECT AVG(DeliveryTimeDays) AS avg_delivery_time
FROM sales_data;

9. Sales by City
SELECT City, SUM(TransactionAmount) AS city_sales
FROM sales_data
GROUP BY City
ORDER BY city_sales DESC;

10.Sales by Region
SELECT Region, SUM(TransactionAmount) AS region_sales
FROM sales_data
GROUP BY Region
ORDER BY region_sales DESC;

11.Sales by Product
SELECT ProductName, SUM(TransactionAmount) AS product_sales
FROM sales_data
GROUP BY ProductName
ORDER BY product_sales DESC;

12.Returned Products Count
SELECT ProductName, COUNT(*) AS returned_count
FROM sales_data
WHERE Returned = 'Yes'
GROUP BY ProductName
ORDER BY returned_count DESC;

13.Sales by Promotion
SELECT IsPromotional, SUM(TransactionAmount) AS promotional_sales
FROM sales_data
GROUP BY IsPromotional
ORDER BY promotional_sales DESC;

14.Customer Age Distribution (Basic Stats)
SELECT MIN(CustomerAge) AS min_age, MAX(CustomerAge) AS max_age, AVG(CustomerAge) AS avg_age,
       PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY CustomerAge) AS age_25_percentile,
       PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY CustomerAge) AS age_50_percentile,
       PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY CustomerAge) AS age_75_percentile
FROM sales_data;

15.Customer Gender Distribution
SELECT CustomerGender, COUNT(*) AS gender_count
FROM sales_data
GROUP BY CustomerGender
ORDER BY gender_count DESC;
